### Visualize, analyze and export the data obtained from a kymograph performed in a C-Trap®

This notebook contains widgets to make it interactive. However, in GitHub, it renders as a static HTML file. Just download the file to interact with your data.

```python
import processes as prs
%matplotlib notebook
```

```python
filepath = r'20200615-173301 Kymograph 27.h5'
```

#### Visualize the kymograph

```python
kymo = prs.plot_Kymograph(filepath, channel = 'green', figsize = (9,5),
                         axis = False, colorbar = True, threshold = 80, 
                         force_correlation = True, downsampled_rate = 220)
```

#### Photon count profile for a particular time

```python
prs.plot_PhotonProfile(filepath, channels = ['green', 'blue'], threshold = 80, show_CSVbuttons = True)
```

#### Export the Photon Counts Matrix

```python
prs.export_PhotonsKymo(filepath, channel = 'all')        # channel = 'red', 'green', 'blue' or 'all'
```
