## SCANS

This notebook allows you to visualize and export the data from a confocal scan performed in the C-Trap.

```python
import processes as prs
%matplotlib notebook
```

```python
filepath = r'20201123-135132 Scan 24.h5'
```

#### Visualize the scan

```python
fig = prs.plot_Scan(filepath, figsize = (8,4), channel = 'green', threshold = 70,
                    force_correlation = True, downsampled_rate = 50,
                    axis = False, showSaveButtons = True)
```

#### Export Photon Count Matrix

```python
prs.export_PhotonsScan(filepath, channel = 'green', frame = 10)
```

#### Export Force as CSV

```python
prs.export_ScanForce(filepath, downsampled_rate=100)
```
