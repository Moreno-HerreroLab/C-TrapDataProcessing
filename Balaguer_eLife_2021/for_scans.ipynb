### Visualize and export the data obtained from a confocal scan performed in a C-Trap® .

NOTE: This notebook contains widgets to make it interactive. However, in GitHub, it renders as a static HTML file. Download the file and run it to be able to interact with your data.

```python
import processes as prs
%matplotlib notebook
```

```python
filepath = r'20201123-135132 Scan 24.h5' # path to your file
```

#### Visualize the scan

```python
fig = prs.plot_Scan(filepath, figsize = (8,4), channel = 'green', threshold = 60,
                    force_correlation = True, downsampled_rate = 50,
                    axis = False, showSaveButtons = True)
```

#### Export Photon Count Matrix

The exported photon counts matrix is stored by default in the same folder as your file.

```python
prs.export_PhotonsScan(filepath, channel = 'all', frame = 10)
```

#### Export Force as CSV

The exported data is stored by default in the same folder as your file.

```python
prs.export_ScanForce(filepath, downsampled_rate = 100)
```
