## KYMOGRAPHS

This notebook allows you to visualize, analize and export the data from a confocal kymograph performed in the C-Trap.

```python
import processes as prs
%matplotlib notebook
```

```python
filepath = r'20200923-133947 Kymograph 18.h5'
```

#### Visualize the kymograph

```python
kymo = prs.plot_Kymograph(filepath, channel = 'green', figsize = (7,5),
                         axis = False, colorbar = True, threshold = None, 
                         force_correlation = True, downsampled_rate = 220)
```

#### Photon count profile for a particular time

```python
prs.plot_PhotonProfile(filepath, channels = ['green', 'blue'], threshold = None, show_CSVbuttons = True)
```

#### Export the Photon Counts Matrix

```python
prs.export_PhotonsKymo(filepath, channel = 'all')        #'red', 'green', 'blue' or 'all'
```
